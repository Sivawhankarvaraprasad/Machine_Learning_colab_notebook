```python
#step1 : import library
import pandas as pd
```

```python
#step2: import /read data set
House = pd.read_csv('https://raw.githubusercontent.com/YBI-Foundation/Dataset/main/Boston.csv')
```

```python
House.head()
```

```python
House.info()
```

```python
House.describe()
```

```python
House.columns
```

```python
#step3 : define target (y) and features(x)
y=House ['MEDV']
x=House[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
#step4 : train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=2529)
x_train
```

```python
#step5: select the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#step 6: train model (fit model)
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
#slop
model.coef_
```

```python
#step7 : prediction
y_predict = model.predict(x_test)
```

```python
#step 8: accuracy
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_predict)
```
