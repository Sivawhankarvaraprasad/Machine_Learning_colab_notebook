**Project Objectives:**

**Sales Prediction:** Build a predictive model to accurately forecast sales of products in different BigMart outlets, leveraging historical sales data and product/outlet characteristics.

**BigMart Sales Dataset**

**Description:** This dataset contains historical sales data for various products sold across different BigMart outlets. It includes information on product attributes, outlet details, and sales figures.

This dataset will be used to train and evaluate the sales prediction model, providing valuable insights for optimizing BigMart's business operations.

```python
#step 3: define y and x
y=sales['Item_Outlet_Sales']
```

```python
y.shape
```

```python
y
```

```python
x=sales[['Item_Identifier', 'Item_Weight', 'Item_Fat_Content', 'Item_Visibility',
       'Item_Type', 'Item_MRP', 'Outlet_Identifier',
       'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type',
       'Outlet_Type']]
```

```python
x=sales.drop(['Item_Identifier','Item_Outlet_Sales'], axis=1)
```

```python
x.shape
```

```python
x
```

```python
#Data Preprocessing
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_std=sales[['Item_Weight', 'Item_Visibility','Item_MRP','Outlet_Establishment_Year']]
```

```python
x_std=sc.fit_transform(x_std)
```

```python
x_std
```

```python
x[['Item_Weight', 'Item_Visibility','Item_MRP','Outlet_Establishment_Year']]=pd.DataFrame(x_std,columns=[['Item_Weight', 'Item_Visibility','Item_MRP','Outlet_Establishment_Year']])
```

```python
x
```

```python
#step4 :split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=2529)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
#step6  : model train
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(random_state=2529)
```

```python
# Convert categorical features to numerical using one-hot encoding
x_train = pd.get_dummies(x_train)
x_test = pd.get_dummies(x_test)
```

```python
# Align columns in training and testing data (if necessary)
x_train, x_test = x_train.align(x_test, join='left', axis=1, fill_value=0)

rfr.fit(x_train,y_train )
```

```python
#step7 : model prediction
y_pred=rfr.predict(x_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
#step8: accuracy
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
